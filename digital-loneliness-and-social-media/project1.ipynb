```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# load dataset
path = "mentalhealth.csv"

# colours
primary = "#351F7D"
secondary = "#C4A3CE"

gradient = [
    "#C4A3CE",
    "#B18CC4",
    "#9D75BA",
    "#895EB0",
    "#7647A5",
    "#62309B",
    "#4F1990",
    "#351F7D"
]
```

exploration of data

```python
file = pd.read_csv(path)
```

```python
print("\nFirst 5 records:")
file.head()
```

```python
print("\nLast 5 records:")
file.tail()
```

```python
print("Dataset description:")
file.describe(include="all")
```

Average Teenager in this dataset

```python
file.describe(include = "all")
```

```python
avgdata = {
"Average Age": file["age"].mean(),
"Average Sleep Hours": file["sleep_hours"].mean(),
"Average Daily Social Media Hours": file["daily_social_media_hours"].mean(),
"Average Physical Activity Hours": file["physical_activity"].mean()
}

df = pd.DataFrame(avgdata, index = ["Dataset Average"])
df.round(2)
```

```python
file["platform_usage"].value_counts()
```

```python
palette = []
plt.figure(figsize=(10, 6))
ax = sns.countplot(y="platform_usage", data=file, order=file["platform_usage"].value_counts().index, palette = gradient, hue = "platform_usage", legend=False)


for container in ax.containers:
    ax.bar_label(container, padding=4)

plt.title("Frequently used Social Media Platforms",fontsize=16)
plt.xlabel("Number of Teenagers")
plt.ylabel("Platform")
plt.tight_layout()
plt.show()
```

```python
file["daily_social_media_hours"].describe()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(file["daily_social_media_hours"], bins=15, kde=True, color="#9E62BB")

plt.title("Distribution of Daily Social Media Usage", fontsize=16)
plt.xlabel("Hour per Day")
plt.ylabel("Number of Teenagers")
plt.tight_layout()
plt.show()
```

```python
file["age_group"] = pd.cut(file["age"], bins=[12, 15, 18, 21], labels=["13-15", "16-18", "19-21"])
file.groupby("age_group")["daily_social_media_hours"].mean().round(2)
```

```python
ageuse = file.groupby("age_group")["daily_social_media_hours"].mean()

plt.figure(figsize=(7,4))
ax = sns.barplot(x=ageuse.index, y=ageuse.values, hue=ageuse.index, palette=gradient[-3:],legend=False)

for container in ax.containers:
    ax.bar_label(container, fmt="%.2f", fontsize=7)

plt.title("Average Daily Social Media Usage by Age Group", fontsize=14)

plt.xlabel("Age Group")
plt.ylabel("Average Hours")

plt.tight_layout()
plt.show()
```

```python
file.groupby("gender")["daily_social_media_hours"].mean().round(2)
```

```python
sns.boxplot(x="gender", y="daily_social_media_hours", data=file, palette=gradient[3:4], hue="gender", legend=False)
plt.title("Social Media Usage by Gender", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
platform_hours = file.groupby("platform_usage")["daily_social_media_hours"].mean().sort_values(ascending=False)
platform_hours.round(2)
```
